### *Test different timepoint weighting strategies*

Here we demonstrate three approaches to decoder recalibration based on HMM outputs:

- equal weighting of all timepoints
- removing low confidence timepoints (where HMM guess for most likely state at that timepoint is below a threshold)
- weighting time points by the square of the HMM's confidence: $P(\text{most likely state at timestep} | \text{data})^2$

We'll use pairs of sessions up to 14 days apart to compare the approaches on. We've subselected session pairs with decent decoder performance after just mean recalibration ($R^{2} > 0.1$ on new day) and sessions with at least 3 blocks of data -  this minimizes issues with strategy changes, weird nonstationarities that can appear, and small dataset sizes.


We'll also use the optimal threshold value (and associated hyperparameters) from a sweep to set up the HMM. This theoretically gives some advantage to the threshold-based approach already:

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
import copy, glob, sys, joblib, argparse
from joblib import Parallel, delayed

import os
HOME = os.path.expanduser('~')

[sys.path.append(f) for f in glob.glob(HOME + '/projects/nonstationarities/utils/*')]
from preprocess import DataStruct
from plotting_utils import figSize
from lineplots import plotsd
from session_utils import *
from recalibration_utils import *
from click_utils import *

from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import FactorAnalysis, PCA
from sklearn.model_selection import ParameterGrid
from hmm_utils import HMMRecalibration


import sweep_utils



# %%%%%%%%%%%% Configurable parameters %%%%%%%%%%%%%%%%%%

gridSize     = 20     # already got specified when making the datasets (possible coordinate diffs across sessions)
stayProb     = 0.999

sweepOpts = dict()
sweepOpts['probWeighted'] = ['probWeighted', 0., 0.2, 0.4, 0.6, 0.8]



# load dataset, add files as a sweep parameter:
DATA_DIR    = '/oak/stanford/groups/shenoy/ghwilson/nonstationarities/T5/train/'
files       = glob.glob(DATA_DIR + '*')
sweepOpts['file'] = files


# generate unchanging arguments for HMM 
nStates = gridSize**2
baseOpts                 = dict()
baseOpts['kappa']        = 2
baseOpts['inflection']   = 50
baseOpts['exp']          = 0.1
baseOpts['pStateStart']  = np.zeros((nStates,1)) + 1/nStates
baseOpts['stateTrans']   = np.eye(nStates)*stayProb 
baseOpts['gridSize']     = gridSize

for x in range(nStates):
    idx                            = np.setdiff1d(np.arange(nStates), x)
    baseOpts['stateTrans'][x, idx] = (1-stayProb)/(nStates-1)
    
    
    

if __name__ == '__main__':
    np.random.seed(42)
    
    # generate sweep args
    sweep_args = sweep_utils.generateArgs(sweepOpts, baseOpts)

    print('Number of jobs: ', len(sweep_args))
    print('Number of CPUs: ', joblib.cpu_count())
    print('Running...')
    
    scores = Parallel(n_jobs= 10, verbose = 0)(delayed(sweep_utils.test_HMM)(arg) for arg in sweep_args)
    
scores_df = pd.DataFrame(scores)
```

```python
import re
import seaborn as sns 


def makeImprovementPlot(df):

    pearson_scores = list()
    r2_scores      = list()

    for weighting in sweepOpts['probWeighted']:
        pearson = df[df['probWeighted'] == weighting]['pearson_r']
        r2      = df[df['probWeighted'] == weighting]['R2_score']
        pearson_scores.append(pearson.to_numpy())
        r2_scores.append(r2.to_numpy())
        
    figSize(5, 15)

    plt.subplot(1, 2, 1)
    plt_data = [pearson_scores[0] - x for i, x in enumerate(pearson_scores) if i > 0]
    sns.boxplot(data = plt_data, orient= 'h', palette = 'Blues')
    plt.axvline(0, color = 'k', linestyle = '--', alpha = 0.4)
    plt.title('Relative pearson improvement')

    plt.subplot(1, 2, 2)
    plt_data = [r2_scores[0] - x for i, x in enumerate(r2_scores) if i > 0]
    sns.boxplot(data = [x[x < 1e20] for x in plt_data], orient= 'h', palette = 'Blues')
    plt.axvline(0, color = 'k', linestyle = '--', alpha = 0.4)
    plt.title('Relative R^2 improvement')

    
    return pearson_scores, r2_scores


def getConsecutivePairs(scores_df):
    
    pat             = '\d\d\d\d\.\d\d\.\d\d'
    unique_dates    = np.unique(np.concatenate([re.findall(pat, x) for x in scores_df['file'] ]))
    unique_pairs    = [(unique_dates[i], unique_dates[i+1]) for i in range(0, len(unique_dates) - 1, 2)]
    unique_patterns = [ x + '_to_' + y for (x, y) in unique_pairs]
    
    consecutive_scores_df = scores_df[scores_df['file'].str.contains('|'.join(unique_patterns))]

    return consecutive_scores_df
                       
```

```python

pearson_scores, r2_scores = makeImprovementPlot(scores_df)

fig_path    = '/home/users/ghwilson/projects/nonstationarities/figures/'

#plt.savefig(fig_path + 'HMM/prob_weighting_results', format = 'pdf')

plt.show()
```

```python
for i in range(5):
    pearson_pval = scipy.stats.wilcoxon(pearson_scores[i + 1], pearson_scores[0])[1]
    r2_pval      = scipy.stats.wilcoxon(r2_scores[i + 1], r2_scores[0])[1]
    
    print('----------')
    print(sweepOpts['probWeighted'][0], ' vs threshold = ', sweepOpts['probWeighted'][i + 1] )
    print('pearson improvement: {} \n R2 improvement: {}'.format(pearson_pval, r2_pval))
```

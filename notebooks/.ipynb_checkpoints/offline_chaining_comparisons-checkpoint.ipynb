### train linear regression on day 0, get baseline performance measure


```python
def generateStretches(files, n_sessions, min_length, overlap = 0, max_gap = np.inf):
    
    # get dates first:
    pat    = '\d\d\d\d\.\d\d\.\d\d'
    dates = [re.findall(pat, f)[0] for f in files]
        
    # now iterate through list, adding to segment until we hit max # or a gap:
    i = 0
    stretches       = []
    current_stretch = []
    while i != len(files):
        
        isSmallStretch = len(current_stretch) < n_sessions
        isEmpty        = len(current_stretch) == 0
        
        isSmallGap     = daysBetween(dates[i], current_stretch[-1]) < max_gap if not isEmpty else False
        
        if isSmallStretch and (isEmpty or isSmallGap):
            current_stretch.append(files[i])
        else:            
            if len(current_stretch) > min_length:
                stretches.append(np.asarray(current_stretch))
            if not isSmallStretch and isSmallGap:
                i -= overlap
            current_stretch = [files[i]]

        i += 1
    
    return stretches 


def runStretch(stretch, cfg, hmm_cfg, stab_cfg):
    
    nStates         = hmm_cfg['gridSize']**2
    stateTrans      = np.eye(nStates) * hmm_cfg['stayProb'] #Define the state transition matrix, which assumes uniform transition probability of transitioning to new state

    for x in range(nStates):
        idx                = np.setdiff1d(np.arange(nStates), x)
        stateTrans[x, idx] = (1-hmm_cfg['stayProb'])/(nStates-1)

    pStateStart = np.zeros((nStates,1)) + (1/nStates)
    adjustKappa = lambda dist : 1 / (1 + np.exp(-1 * (dist - hmm_cfg['inflection']) * hmm_cfg['exp']))


    scores    = np.empty((6, len(stretch))) 
    scores[:] = np.nan
    dates     = np.empty(len(stretch), dtype = object)
    # rows are 0 - no recalibration; 1 - mean recalibration; 
    # 2 - full recalibration; 3 - stabilizer recalibration; 
    # 4 - stabilizer (chained); 5- HMM recalibration (each session)
    fields  = ['TX', 'cursorPos', 'targetPos']

    # Initialize first day
    day0         = DataStruct(stretch[0], alignScreens = True, causal_filter = cfg['sigma'])
    session_dat  = getTrainTest(day0, train_size = cfg['train_size'], task = cfg['task'], blocks = block_constraints[stretch[0]],
                                fields = fields, returnFlattened = True)

    train_y         = np.concatenate(session_dat['train_targetPos']) - np.concatenate(session_dat['train_cursorPos'])
    test_y          = np.concatenate(session_dat['test_targetPos']) - np.concatenate(session_dat['test_cursorPos'])
    train_x, test_x = get_BlockwiseMeanSubtracted(session_dat['train_TX'], session_dat['test_TX'], concatenate = True)
    ref_means       = np.concatenate(session_dat['train_TX']).mean(axis = 0)

    ref_decoder   = LinearRegression(fit_intercept = False, normalize = False).fit(train_x, train_y)
    scores[:, 0]  = ref_decoder.score(test_x, test_y)
    dates[0]      = day0.date

    stab = stabilizer_utils.Stabilizer(model_type = stab_cfg['model_type'], n_components = stab_cfg['n_components'])
    stab.fit_ref(train_x)
    stab_daisychained = copy.deepcopy(stab)

    latent_decoder = LinearRegression(fit_intercept = False, normalize = False).fit(stab.ref_model.transform(train_x), train_y)
    scores[3:5, 0] = latent_decoder.score(stab.ref_model.transform(test_x), test_y)

    models        = list()
    hmm_new_ref   = deepcopy(ref_decoder)
    prior_decoder = deepcopy(ref_decoder)
    
    for i, file in enumerate(stretch[1:]):
        new_dat    = DataStruct(file, alignScreens = True, causal_filter = cfg['sigma'])

        session_dat       = getTrainTest(new_dat, train_size = cfg['train_size'], task = cfg['task'], blocks = block_constraints[file], fields = fields, returnFlattened = True)
        train_y           = np.concatenate(session_dat['train_targetPos']) - np.concatenate(session_dat['train_cursorPos'])
        test_y            = np.concatenate(session_dat['test_targetPos']) - np.concatenate(session_dat['test_cursorPos'])
        train_x, test_x   = subtractMeans(session_dat['train_TX'], session_dat['test_TX'], 
                                          method = 'blockwise', concatenate = True)
        test_x_stalemeans = np.concatenate(session_dat['test_TX']) - ref_means
        cursorPos         = np.concatenate(session_dat['train_cursorPos'])
        
        
        new_decoder       = LinearRegression(fit_intercept = False, normalize = False).fit(train_x, train_y)

        # fit stabilizers
        stab.fit_new(train_x, B = stab_cfg['B'], thresh = stab_cfg['thresh'])
        stab_daisychained.fit_new(train_x, B = stab_cfg['B'], thresh = stab_cfg['thresh'], daisy_chain = True)
        test_latent              = stab.transform(test_x)
        test_latent_daisychained = stab_daisychained.transform(test_x)
        
        # fit HMM 
        targLocs      = hmm_utils.get_DiscreteTargetGrid(new_dat, gridSize = hmm_cfg['gridSize'], task = cfg['task'])
        hmm           = HMMRecalibration(stateTrans, targLocs, pStateStart, hmm_cfg['vmKappa'], adjustKappa)
        hmm_new_ref   = hmm.recalibrate(hmm_new_ref, [train_x], [cursorPos], probThreshold = hmm_cfg['probThreshold'])

        suprecal_score  = new_decoder.score(test_x, test_y)
        meanrecal_score = prior_decoder.score(test_x, test_y)
        
        if suprecal_score < cfg['min_suprecal_R2'] or (cfg['min_meanrecal_R2'] is not None and meanrecal_score <  cfg['min_meanrecal_R2']):
            break
            
        scores[0, i+1] = ref_decoder.score(test_x_stalemeans, test_y) # no recalibration
        scores[1, i+1] = meanrecal_score               # mean recalibration
        scores[2, i+1] = suprecal_score                # full recalibration

        scores[3, i+1] = latent_decoder.score(test_latent, test_y) 
        scores[4, i+1] = latent_decoder.score(test_latent_daisychained, test_y)
        scores[5, i+1] = hmm_new_ref.score(test_x, test_y)

        prior_decoder = deepcopy(new_decoder)
        
        dates[i+1] = new_dat.date

        print(new_dat.date, ' finished')
    
    return scores, dates 
```

```python
stretches = generateStretches(files, n_sessions = 10, min_length = 5, overlap = 7, max_gap = np.inf)

for stretch in stretches:
    pat   = '\d\d\d\d\.\d\d\.\d\d'
    dates = [re.findall(pat, f)[0] for f in stretch]
    [print(d) for d in dates]
    print('\n')
```

```python
from joblib import Parallel, delayed
import time 

# General parameters: 
cfg = {'train_size' : 0.5, 'task' : None, 'sigma' : 2, 
       'min_meanrecal_R2' : 0.15, 'min_suprecal_R2' : 0.15}

# recalibration settings:
hmm_cfg  = {'gridSize' : 20, 'stayProb' : 0.999, 'vmKappa' : 2, 
            'inflection' : 70, 'exp' : 0.5, 'probThreshold' : 'probWeighted'}
stab_cfg = {'model_type' : 'FactorAnalysis', 'n_components' : 6, 'B' : 160, 'thresh' : 0.01}

parallel_out = Parallel(n_jobs= - 1, verbose = 5)(delayed(runStretch)(stretch, cfg, hmm_cfg, stab_cfg) for stretch in stretches)
```

```python
scores_list = []
dates_list = []

for (scores, dates) in parallel_out:
    if len(dates[dates != None]) > 2:
        scores_list.append(scores)
        dates_list.append(dates)
```

```python
import matplotlib as mpl
import pandas as pd

n_days = 600



df_list      = list()
fence_colors = list()
for j in range(6):
    data    = np.empty((n_days, len(scores_list)))
    data[:] = np.nan
    idx     = 0
    for scores, dates in zip(scores_list, dates_list):
        
        not_none = dates != None
        
        dates  = dates[not_none]
        scores = scores[:, :len(dates)]
        days   = np.asarray([daysBetween(dates[0], dates[i]) for i in range(len(dates))])
        
        days   = days[days < n_days]
        scores = scores[:, :len(days)]
        
        data[days, idx] = scores[j, :] / scores[4, :]

        c = mpl.colors.to_rgba(colors[j], alpha = 0.2)
        fence_colors.append(c)
        idx += 1
    
    data     = pd.DataFrame(data = data.T).interpolate('linear', axis = 1)
    sort_idx = np.argsort(np.nansum(data, axis = 1))
    data     = data.iloc[sort_idx, :]
    df_list.append(data)

interpolated_df = pd.concat(df_list, axis = 0).reset_index()
```

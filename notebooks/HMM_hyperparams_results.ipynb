Code below does a grid search for optimal HMM values:


```python
%load_ext autoreload
%autoreload 2

import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
import glob
import sys

[sys.path.append(f) for f in glob.glob('../utils/*')]
import preprocess, sweep_utils
from plotting_utils import figSize, setTickLabels
from lineplots import plotsd
from session_utils import *
from recalibration_utils import *
from click_utils import *


fig_path  = '/home/users/ghwilson/projects/nonstationarities/figures/'
SWEEP_DIR = '/oak/stanford/groups/shenoy/gwilson/nonstationarities/T5/HMM/HMMsweep/'
params    = ['kappa', 'inflection', 'exp']


files     = glob.glob(SWEEP_DIR + '*')
scores_df = sweep_utils.getSummaryDataFrame(files, params + ['R2_score', 'pearson_r', 'days_apart', 'file'])
```

```python
params      = ['kappa', 'inflection', 'exp']
unique_vals = dict([(x, np.unique(scores_df[x])) for x in params])
unique_lens = [len(np.unique(scores_df[x])) for x in params]
nSamples    = int(scores_df.shape[0] / np.product([len(x) for x in unique_vals.values()]))
scores_arr  = np.zeros([*unique_lens, nSamples])

lens = list()
for i, kappa in enumerate(unique_vals['kappa']):
    for j, inflection in enumerate(unique_vals['inflection']):
        for k, exp in enumerate(unique_vals['exp']):
            subset                 = scores_df.loc[(scores_df["kappa"] == kappa) & (scores_df["inflection"] == inflection) & (scores_df["exp"] == exp)]
            lens.append(len(subset))
            scores_arr[i, j, k, :] = subset['R2_score']
```

```python
figSize(5, 20)

medscores = np.median(scores_arr, axis = -1)

opt_kappa_idx, opt_inflection_idx, opt_exp_idx = np.unravel_index(medscores.argmax(), medscores.shape)

opt_kappa      = unique_vals['kappa'][opt_kappa_idx]
opt_inflection = unique_vals['inflection'][opt_inflection_idx]
opt_exp        = unique_vals['exp'][opt_exp_idx]


opt_dict = dict()
opt_dict['kappa']        = opt_kappa
opt_dict['inflection']   = opt_inflection
opt_dict['exp']          = opt_exp

clim = [0, 0.35]


plt.subplot(1, 3, 1)
plt.imshow(medscores[opt_kappa_idx, :, :], aspect = 'auto')
plt.title('Varying exponent, inflection (fixed kappa)', fontweight = 'bold')
setTickLabels(plt.gca(), unique_vals['inflection'], 'y')
setTickLabels(plt.gca(), unique_vals['exp'], 'x')
plt.xlabel('Exponent')
plt.ylabel('Inflection')
plt.clim(clim)

plt.subplot(1, 3, 2)
plt.imshow(medscores[:, opt_inflection_idx, :], aspect = 'auto')
plt.title('Varying kappa, inflection (fixed inflection)', fontweight = 'bold')
setTickLabels(plt.gca(), unique_vals['kappa'], 'y')
setTickLabels(plt.gca(), unique_vals['exp'], 'x')
plt.xlabel('Exponent')
plt.ylabel('Kappa')
plt.clim(clim)

plt.subplot(1, 3, 3)
plt.imshow(medscores[:, :, opt_exp_idx], aspect = 'auto')
plt.title('Varying kappa, exponent (fixed exponent)', fontweight = 'bold')
plt.clim(clim)
setTickLabels(plt.gca(), unique_vals['kappa'], 'y')
setTickLabels(plt.gca(), unique_vals['inflection'], 'x')
plt.xlabel('Inflection')
plt.ylabel('Kappa')
cbar = plt.colorbar()

cbar.set_label('median $R^2$', fontweight = 'medium', rotation = 270, labelpad = 20)
plt.savefig(fig_path + 'HMM/T5_hyperparam_gridplots.pdf')

plt.show()
```

```python
def plotTrials(results, n_plot =  10):
    n_trls = len(results['trialStart'])

    starts = np.concatenate([results['trialStart'], [results['posTraj'].shape[0] - 1]])
    for j in np.random.choice(n_trls, min(n_plot, n_trls), replace = False):
        start = starts[j]
        stop  = starts[j + 1]
        targ        = results['targTraj'][stop, :] 
        trl_cursor  = results['posTraj'][start:stop, :] - results['posTraj'][start, :]

        _ = plt.scatter(trl_cursor[:, 0], trl_cursor[:, 1], color = 'k', alpha = 0.1)
```

```python
from sklearn.decomposition import PCA, FactorAnalysis
from CosineTuning import getAngles, angle_between


def getLatentsAndAngles(results):
    
    #pca     = PCA(n_components = 2).fit(results['neuralTraj'])
    pca     = FactorAnalysis(n_components = 2).fit(results['neuralTraj'])
    latents = pca.transform(results['neuralTraj'])
    angles  = getAngles(results['velTraj'], returnFullAngle = True)
    
    return latents, pca

def getDirectionVectors(pca, cfg):
    
    direction_vectors  = pca.components_.dot(cfg['neuralTuning'][:, 1:])
    direction_vectors /= np.linalg.norm(direction_vectors, axis = 0)
    
    return direction_vectors
```

```python
figSize(5, 20)

nIters = 4

for i in range(nIters):
    
    plt.subplot(1, 4, i + 1)
    cfg = initializeBCI(base_opts)
    
    cfg_gravity            = copy.deepcopy(cfg)
    cfg_gravity['gravity'] = np.asarray([[0, -0.35]]).T

    cfg_gravity['beta'] = gainSweep(cfg_gravity, base_opts['possibleGain'])
    results_gravity     = simulateBCIFitts(cfg_gravity) 
    
    latents_gravity, pca_gravity = getLatentsAndAngles(results_gravity)
    sns.kdeplot(x   = latents_gravity[:, 0], y = latents_gravity[:, 1], fill = True)

    direction_vectors = getDirectionVectors(pca_gravity, cfg_gravity)
    plt.quiver([0, 0], [0, 0], direction_vectors[0, :], direction_vectors[1, :],
           color = ['r', 'g'], units = 'xy', scale = 1)

    skew = scipy.stats.skew(latents_gravity, axis = 0)
    plt.quiver(0, 0, *skew, color = 'orange', units = 'xy', scale = 1)
    
    horz_angle = angle_between(skew, direction_vectors[:, 0], returnFullAngle = True) * np.sign(np.cross(skew, direction_vectors[:, 0]))
    vert_angle = angle_between(skew, direction_vectors[:, 1], returnFullAngle = True) * np.sign(np.cross(skew, direction_vectors[:, 1]))

    
    print(f'{i + 1}/{nIters}')
    

fname = os.path.join(fig_path, 'simulator/example_ablebodied_subspaces')
plt.savefig(fname, format = 'pdf')
    
plt.show()
```

```python
def measureSkewAngles(base_opts, gravity = False):
    cfg = initializeBCI(base_opts)
    
    cfg_gravity            = copy.deepcopy(cfg)
    if gravity:
        cfg_gravity['gravity'] = np.asarray([[0, -0.35]]).T

    cfg_gravity['beta'] = gainSweep(cfg_gravity, base_opts['possibleGain'])
    results_gravity     = simulateBCIFitts(cfg_gravity) 
    
    latents_gravity, pca_gravity = getLatentsAndAngles(results_gravity)
    direction_vectors            = getDirectionVectors(pca_gravity, cfg_gravity)
    skew                         = scipy.stats.skew(latents_gravity, axis = 0)
    
    horz_angle = angle_between(skew, direction_vectors[:, 0], returnFullAngle = True) * np.sign(np.cross(skew, direction_vectors[:, 0]))
    vert_angle = angle_between(skew, direction_vectors[:, 1], returnFullAngle = True) * np.sign(np.cross(skew, direction_vectors[:, 1]))
    
    return horz_angle, vert_angle
```

```python
from joblib import Parallel, delayed

nIters = 500

rep_data = Parallel(n_jobs= -1, verbose = 5)(delayed(measureSkewAngles)(base_opts, gravity = False) for i in range(nIters))
rep_data_gravity = Parallel(n_jobs= -1, verbose = 5)(delayed(measureSkewAngles)(base_opts, gravity = True) for i in range(nIters))
```

```python
def circular_hist(ax, x, bins=16, density=True, offset=0, gaps=True, edgecolor = 'C0'):
    """
    Produce a circular histogram of angles on ax.

    Parameters
    ----------
    ax : matplotlib.axes._subplots.PolarAxesSubplot
        axis instance created with subplot_kw=dict(projection='polar').

    x : array
        Angles to plot, expected in units of radians.

    bins : int, optional
        Defines the number of equal-width bins in the range. The default is 16.

    density : bool, optional
        If True plot frequency proportional to area. If False plot frequency
        proportional to radius. The default is True.

    offset : float, optional
        Sets the offset for the location of the 0 direction in units of
        radians. The default is 0.

    gaps : bool, optional
        Whether to allow gaps between bins. When gaps = False the bins are
        forced to partition the entire [-pi, pi] range. The default is True.

    Returns
    -------
    n : array or list of arrays
        The number of values in each bin.

    bins : array
        The edges of the bins.

    patches : `.BarContainer` or list of a single `.Polygon`
        Container of individual artists used to create the histogram
        or list of such containers if there are multiple input datasets.
    """
    # Wrap angles to [-pi, pi)
    x = (x+np.pi) % (2*np.pi) - np.pi

    # Force bins to partition entire circle
    if not gaps:
        bins = np.linspace(-np.pi, np.pi, num=bins+1)

    # Bin data and record counts
    n, bins = np.histogram(x, bins=bins)

    # Compute width of each bin
    widths = np.diff(bins)

    # By default plot frequency proportional to area
    if density:
        # Area to assign each bin
        area = n / x.size
        # Calculate corresponding bin radius
        radius = (area/np.pi) ** .5
    # Otherwise plot frequency proportional to radius
    else:
        radius = n

    # Plot data on ax
    patches = ax.bar(bins[:-1], radius, zorder=1, align='edge', width=widths,
                     edgecolor=edgecolor, color = edgecolor, fill=True, alpha = 0.8,
                     linewidth=1)

    # Set the direction of the zero angle
    ax.set_theta_offset(offset)

    # Remove ylabels for area plots (they are mostly obstructive)
    if density:
        ax.set_yticks([])

    return n, bins, patches
```

```python
figSize(5, 10)


for i, angles in enumerate(([np.vstack(rep_data), np.vstack(rep_data_gravity)])):
    
    plt.subplot(1, 2, i + 1, projection = 'polar')
    ax = plt.gca()

    circular_hist(ax, angles[:, 0], edgecolor = 'r')
    circular_hist(ax, angles[:, 1], edgecolor = 'g')
    plt.legend(['angle with horizontal', 'angle with vertical'])

fname = os.path.join(fig_path, 'simulator/gravity_angle_distributions')
plt.savefig(fname, format = 'pdf')

plt.show()
```

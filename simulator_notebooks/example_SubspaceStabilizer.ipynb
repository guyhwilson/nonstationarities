### **Example of subspace realignment in simulator**

We first learn some map $f: X \rightarrow Z \rightarrow Y $ where:

- $X$ is neural activity on day A
- $Y$ is behavior on day A
- $Z$ is some latent space activity on day A

this map is a composition of two linear functions $g: X \rightarrow Z$ (found via linear dimensionality reduction) and $h: Z \rightarrow Y $ (via regression against behavior). On a new day B, we update $g$ while keeping $h$ fixed. To deal with the difference between these subspaces, we now learn a mapping between them via some orthogonal transformation $Q$. Yielding:

$$f': X' \rightarrow Y, f' = h \circ (Q \circ g') $$

We can view this as a form of reduced-rank regression (RRR), where the bottleneck is determined by unsupervised learning instead of the RRR loss function.

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import sys, glob
[sys.path.append(f) for f in glob.glob('../utils/*')]
from plotting_utils import figSize
from simulation import *
import simulation_utils

# for a reproducible result
np.random.seed(1)
```

#### Setup


First do a sweep of factor analysis latent dimensionality values 

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from CosineTuning import getAngles
import matplotlib.cm as cm

figSize(5, 12)

nUnits         = 192
SNR            = 0.5
nTrainingSteps = 10000


#The first column of initialTuning is the means, and the second two columns are the preferred directions. 
#We make sure that the last two columns are orthogonal (uniformly distributed PDs)
initialTuning       = simulation_utils.generateUnits(n_units = nUnits, SNR = SNR)
initialTuning[:, 0] = 0  # for now - assume features are centered 

calNeural, calVelocity = simulation_utils.simulateUnitActivity(initialTuning, noise = 0.3, nSteps = nTrainingSteps)

# split into train/test and generate latent trajectories:
train_x, test_x, train_y, test_y = train_test_split(calNeural, calVelocity, train_size = 0.7, random_state = 42)

fa            = FactorAnalysis().fit(train_x)
train_latent  = fa.transform(train_x)
test_latent   = fa.transform(test_x)

# sweep through number of components used and measure holdout R^2:
scores  = np.zeros((20,))
for i in range(20):
    lr        = LinearRegression(fit_intercept = True).fit(train_latent[:,:(i+1)], train_y)
    scores[i] = lr.score(test_latent[:, :(i+1)], test_y)

ref_score = LinearRegression(fit_intercept = True).fit(train_x, train_y).score(test_x, test_y)

# plot performance as function of # components:
plt.subplot(1, 2, 1)
plt.plot(scores[0:], c = 'r')
plt.axhline(ref_score, c = 'k', linestyle = '--')
plt.xlabel("Number of components")
plt.ylabel('R^2')

opt_n = np.argmax(scores) + 1
print('Optimal latent dimensionality: ', opt_n)

# now plot top two weightings and color by angle at timestep:
plt.subplot(1, 2, 2)
angles = getAngles(test_y, returnFullAngle = True)

plt.scatter(test_latent[:, 0], test_latent[:, 1], c = angles, cmap = cm.hsv)
plt.title('Projected neural activity (top two components)')
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Angle (radians)', rotation = 270)
plt.tight_layout()
plt.show()
```

Define code for pulling out matrix mapping to latent space (and test that it's working)

```python
import stabilizer_utils
    
Q = stabilizer_utils.get_FactorAnalysisMap(fa)

x = fa.transform(train_x)
y = (train_x - train_x.mean(axis = 0)).dot(Q)

assert np.linalg.norm(x - y) < 1e-10, print("uh oh")
```

Now fit a subspace-based regression model and test on a given day:

```python
from stabilizer_utils import Stabilizer

model_type = 'FactorAnalysis'
opt_n      = 4

#--------------------
stab         = Stabilizer(model_type = model_type, n_components = opt_n)
stab.fit_ref([train_x])

Q_ref        = stabilizer_utils.get_FactorAnalysisMap(stab.ref_model)
train_latent = (train_x - train_x.mean(axis = 0)).dot(Q_ref)
lr           = LinearRegression(fit_intercept = True).fit(train_latent, train_y)

h      = lr.coef_                                             # map from latent space onto targets
D_coef = h.dot(Q_ref.T)                                       # compose to get neural --> latent --> output 
D      = np.hstack([lr.intercept_[:, np.newaxis], D_coef]).T  # add bias terms

# Normalize the gain of this decoder so that it will output vectors with a magnitude of 1 when the encoded velocity has a magnitude of 1. 
D[:, 0] = D[:,0] / np.linalg.norm(D[1:, :][:, 0]) / np.linalg.norm(initialTuning[:, 1])
D[:, 1] = D[:,1] / np.linalg.norm(D[1:, :][:, 1]) / np.linalg.norm(initialTuning[:, 2])
```

```python
import copy

cfg = dict()
cfg['neuralTuning'] = initialTuning
cfg['D']           = D
cfg['alpha']       = 0.94 # amount of exponential smoothing 
cfg['delT']        = 0.01 # define the time step (10 ms)
cfg['nDelaySteps'] = 20   # define the simulated user's visual feedback delay (200 ms)
cfg['nSimSteps']   = 10000

possibleGain = np.linspace(0.5,2.5,10)
cfg['beta']  = simulation_utils.gainSweep(cfg, possibleGain)

# Simulate BCI performance with matched neural tuning and decoder, and an optimized gain
PD_shrinkage = 0.5
nSimSteps    = 10000

results_orig = simulateBCIFitts(cfg)                                                              
print('With a matched, optimized decoder, mean trial time is ', np.mean(results_orig['ttt']), ' s')

# Simulate a change in neural tuning (specifically, change the PDs only, making sure the total magnitude of tuning is the same)
miscal_cfg = copy.deepcopy(cfg)

miscal_cfg['neuralTuning'] = simulation_utils.simulateTuningShift(initialTuning, n_stable = 0, PD_shrinkage = PD_shrinkage)
results_miscal             = simulateBCIFitts(miscal_cfg)
print('With changed tuning and a mismatched decoder, the mean trial time is ', np.mean(results_miscal['ttt']), ' s')

#Control: increase gain in original tuning subspace 
control_cfg          = copy.deepcopy(cfg)
control_cfg['beta'] /= PD_shrinkage
results_control      = simulateBCIFitts(control_cfg)
print('Control: With changed tuning and a mismatched decoder with restored gain, the mean trial time is ', 
      np.mean(results_control['ttt']), ' s')
```

```python
figSize(5, 10)


from scipy.stats import norm
import seaborn as sns

groups = [x['ttt'] for x in [results_orig, results_miscal, results_control]]

sns.boxplot(data = groups)
sns.swarmplot(data = groups, color = 'k', alpha = 0.5)
ax = plt.gca()
labels = ['Original','Tuning Change','Gain compensation']
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.ylabel('Trial time (s)')
plt.ylim([0, max([max(i) for i in groups]) + 0.2])
plt.title('Time to target with simulated PD shift', fontweight = 'bold')
plt.show()
```

Now modify our original decoder to use the new factor analysis model and stabilizing transformation:

```python
G_new     = get_FactorAnalysisMap(stab.new_model)
D_coefnew = h.dot(G_new.dot(stab.R).T)                                # compose to get neural --> latent --> output 
D_new     = np.hstack([lr.intercept_[:, np.newaxis], D_coefnew]).T    # add bias terms

# Normalize the gain of this decoder so that it will output vectors with a magnitude of 1 when the encoded velocity has a magnitude of 1. 
D_new[:, 0] = D_new[:,0] / np.linalg.norm(D_new[1:, :][:, 0]) / np.linalg.norm(newTuning[:, 1])
D_new[:, 1] = D_new[:,1] / np.linalg.norm(D_new[1:, :][:, 1]) / np.linalg.norm(newTuning[:, 2])
```

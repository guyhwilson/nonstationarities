### Systematic performance comparison of vanilla HMM approach against subspace realignment and mean recalibration:
- use subspace realignment optimal settings from "stability_SubspaceRealignment.ipynb"
- use all pairs of valid sessions

```python
%load_ext autoreload
%autoreload 2

import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from copy import deepcopy
import glob
import sys

[sys.path.append(f) for f in glob.glob('utils/*')]
from preprocess import DataStruct, daysBetween
from firingrate import raster2FR
from plotting_utils import *
from lineplots import plotsd
from hmm_utils import *
from stabilizer_utils import *
from recalibration_utils import *
from session_utils import *


from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import FactorAnalysis, PCA


files          = glob.glob('D:/T5_ClosedLoop/*')

sessions_check = np.load('misc_data/sessions_check.npy', allow_pickle = True).item()
min_nblocks    = 3
max_ndays      = 14
min_R2         = 0.1

init_pairs    = get_SessionPairs(files, min_nblocks = min_nblocks, max_ndays = max_ndays, manually_remove = sessions_check['bad_days'])
pairs, scores = get_StrongTransferPairs(init_pairs, min_R2 = min_R2, train_frac = 0.5, block_constraints = sessions_check)
n_pairs       = len(pairs)
```

```python

# general settings:
np.random.seed(42)
diffs           = list()
task            = 'cursor'
train_frac      = 0.5

# subspace stabilizer settings:
model_type      = 'FactorAnalysis'
n_components    = 10
B               = 160
thresh          = 0.1
sigma           = None

# HMM settings: 
gridSize      = 20  
stayProb      = 0.9999
kappa_sweep   = [0.5, 1, 2, 4, 5, 6]
thresh_sweep  = [0, 0.2, 0.4, 0.5]
#vmKappa       = 2
#probThreshold = 0.5


#--------------------------


nStates       = gridSize**2
stateTrans    = np.eye(nStates)*stayProb #Define the state transition matrix, which assumes uniform transition probability of transitioning to new state

for x in range(nStates):
    idx                = np.setdiff1d(np.arange(nStates), x)
    stateTrans[x, idx] = (1-stayProb)/(nStates-1)
pStateStart = np.zeros((nStates,1)) + 1/nStates



dayA_scores    = np.zeros((3, n_pairs))             # row 0: full decoder (train), row 1: full decoder (test), row 2: latent decoder (test)
dayB_scores    = np.zeros((4, n_pairs))             # rows are 0 - no recalibration; 1 - mean recalibration; 2 - full recalibration; 3 - subspace realignment
dayB_HMMscores = np.zeros((len(kappa_sweep),        # contains HMM recal performance for each hyperparameter value and session pairing
                 len(thresh_sweep), n_pairs))
dates, diffs   = list(), list()                     # keep track of which files used in each pairing (entries are tuples of [<ref file>, <newday file>]) 
diffs          = np.zeros((n_pairs,))               # track the # of days between sessions in each pairing



for i, (A_file, B_file) in enumerate(pairs):  
    dayA, dayB              = DataStruct(A_file), DataStruct(B_file)
    diffs[i]                = daysBetween(dayA.date, dayB.date) # record number of days between sessions

    dayA_blocks             = [sessions_check[A_file] if A_file in sessions_check.keys() else None][0]
    dayB_blocks             = [sessions_check[B_file] if B_file in sessions_check.keys() else None][0] 
    dayA_task, dayB_task, _ = getPairTasks(dayA, dayB, task = task)

     # obtain features and cursorError targets:
    Atrain_x, Atest_x, Atrain_y, Atest_y                 = getTrainTest(dayA, train_frac = train_frac, sigma = sigma, blocks = dayA_blocks, task = dayA_task, return_flattened = True)
    Btrain_x, B_cursorPos, Btrain_y, Btest_x, _, Btest_y = prep_HMMData(dayB, train_frac = train_frac, blocks = dayB_blocks, task = task, return_flattened = True)

    #Btrain_x, Btest_x, Btrain_y, Btest_y = getTrainTest(dayB, train_frac = train_frac, sigma = sigma, blocks = dayB_blocks, task = dayB_task, return_flattened = True)
    
    
    stab                                 = Stabilizer(model_type, n_components)
    stab.fit_ref(Atrain_x, conditionAveraged = False)
    Atrain_latent, Atest_latent          = stab.ref_model.transform(Atrain_x), stab.ref_model.transform(Atest_x)

    stab.fit_new(Btrain_x, B = B, thresh = thresh, conditionAveraged = False, conditions = None)
    Btrain_latent, Btest_latent          = stab.transform(Btrain_x), stab.transform(Btest_x)

    full_score, full_decoder     = traintest_DecoderSupervised([Atrain_x], [Atrain_x], [Atrain_y], [Atrain_y], meanRecal = False)           # evaluate training set mean performance 
    latent_score, latent_decoder = traintest_DecoderSupervised([Atrain_latent], [Atest_latent], [Atrain_y], [Atest_y], meanRecal = False)   # evaluate latent decoder same-day performance 

    dayA_scores[0, i]                     = np.mean(full_score)
    dayA_Scores[1, i]                     = traintest_DecoderSupervised([Atrain_x], [Atest_x], [Atrain_y], [Atest_y], meanRecal = False)[0][0]  # evaluate testing set mean performance 
    dayA_scores[2, i]                     = np.mean(latent_score)

    targLocs                              = get_DiscreteTargetGrid(dayB, gridSize = gridSize, task = dayB_task)

    # ----- REFACTOR BELOW --------

    dayB_scores[0, i] = full_decoder.score(Btest_x - Atrain_x.mean(axis = 0), Btest_y)                                          # no recalibration
    dayB_scores[1, i] = full_decoder.score(Btest_x - Btrain_x.mean(axis = 0), Btest_y)                                          # mean recalibration
    dayB_scores[2, i] = traintest_DecoderSupervised([Btrain_latent], [Btest_latent], [Btrain_y], [Btest_y], meanRecal = False)  # full recalibration                                                    
    dayB_scores[3, i] = latent_decoder.score(Btest_x_latent, Btest_y)    

    for j, vmKappa in enumerate(kappa_sweep):
        for k, probThreshold in enumerate(thresh_sweep):
            hmm_decoder             = train_HMMRecalibrate(deepcopy(ref_decoder), [train_x], [cursorPos], stateTrans, pStateStart, targLocs, vmKappa, probThreshold)                           
            dayB_HMMscores[j, k, i] = hmm_score.(test_x - train_x.mean(axis = 0), test_y)


    
```

```python
figSize(10, 20)

alt_methods = ['mean recal', 'subspace stabilizer']
alt_scores  = [dayB_scores[1, :], dayB_scores[3, :]]

for i in range(2):
    score_diffs    = np.median(dayB_HMMscores - alt_scores[i], axis = 2)
    opt_D, opt_B   = np.unravel_index(score_diffs.argmax(), score_diffs.shape)

    plt.subplot(1, 2, i + 1)
    plt.imshow(score_diffs)
    plt.text(opt_B, opt_D, '*', fontsize = 20)
    plt.ylabel('Kappa', fontweight = 'bold', fontsize = 12)
    plt.xlabel('ProbThresh', fontweight = 'bold', fontsize = 12)
    plt.title('Median pairwise performance difference (HMM vs. ' + alt_methods[i] + ')', fontweight = 'bold')
    plt.colorbar()
    plt.clim([-0.07, 0.07])
    ax = plt.gca()
    _  = ax.set_yticks(np.arange(len(kappa_sweep)))
    _  = ax.set_xticks(np.arange(len(thresh_sweep)))
    _  = ax.set_yticklabels(kappa_sweep,)
    _  = ax.set_xticklabels(thresh_sweep)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import TheilSenRegressor
from resample import bootstrap_LinearRegression

n_bootstraps = 500

# normalize by fully recalibrated performance on new day (upper limit on performance):
subspace_normalized = dayB_scores[3, :]               # / dayB_fullscores[2, :]
HMM_normalized      = dayB_HMMscores[opt_D, opt_B, :] #/ dayB_fullscores[2, :]

lm_ss, means_ss, coefs_ss    = bootstrap_LinearRegression(diffs[:, np.newaxis], subspace_normalized, regressor = TheilSenRegressor(), n_bootstraps = n_bootstraps, random_state = 42)
lm_HMM, means_HMM, coefs_HMM = bootstrap_LinearRegression(diffs[:, np.newaxis], HMM_normalized, regressor = TheilSenRegressor(), n_bootstraps = n_bootstraps, random_state = 42)
```

### check out badly performing days:

```python
figSize(15, 15)

bad_days      = np.where(np.logical_and(diffs < 4, dayB_HMMscores[opt_D, opt_B, :] < 0.1))[0]
good_days     = np.where(np.logical_and(diffs < 4, dayB_HMMscores[opt_D, opt_B, :] > 0.3))[0]
width, height = np.ceil(np.sqrt(len(bad_days))).astype('int'), np.ceil(np.sqrt(len(bad_days))).astype('int')
FR_diffs      = np.zeros((2,  192, len(files) - 1, ))
decoder_preds = list()

for i in range(len(files) - 1):
    dayA = DataStruct(files[i])
    dayB = DataStruct(files[i + 1])

    if sum(dayB.trialType == task) * sum(dayA.trialType == task) != 0:
        dayA_task, dayB_task = 'cursor', 'cursor'   # try to get same task type for both sesions
    else:
        dayA_task = np.unique(dayA.trialType)[0]
        dayB_task = np.unique(dayB.trialType)[0]

    # Fit day A decoders: 
    Atrain_x, Atest_x, Atrain_y, Atest_y = getTrainTest(dayA, train_frac = train_frac, task = dayA_task)
    Btrain_x, Btest_x, Btrain_y, Btest_y = getTrainTest(dayB, train_frac = train_frac, task = dayB_task)
    
    FR_diffs[0, :, i] = Atrain_x.mean(axis = 0)
    FR_diffs[1, :, i] = Btrain_x.mean(axis = 0)
    lm_a              = LinearRegression(fit_intercept = False).fit(Atrain_x - Atrain_x.mean(axis = 0), Atrain_y)
    preds             = lm_a.predict(Btest_x - Btest_x.mean(axis = 0))
    
    decoder_preds.append([Btest_y, preds])
    
```

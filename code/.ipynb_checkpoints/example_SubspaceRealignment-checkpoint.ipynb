```python
# Fit PCA model to trajectories around goCue for radial-8 task:

dat      = DataStruct(files[2])
window   = 5

model_type         = 'PCA'
n_components       = 3
data               = np.dstack([dat.TX[i][:20, :] for i in np.where(dat.trialType == 'cursor')[0] ])
unique, conditions = np.unique(np.vstack([dat.targetPos[i][0, :] for i in np.where(dat.trialType == 'cursor')[0] ]), return_inverse = True, axis = 0)
model, ll          = fit_ConditionAveragedModel(model_type, {'n_components' : n_components}, data, conditions)
```

```python
model_type         = 'FactorAnalysis'
n_components       = 12
data               = np.dstack([dat.TX[i][:20, :] for i in np.where(dat.trialType == 'cursor')[0] ])
unique, conditions = np.unique(np.vstack([dat.targetPos[i][0, :] for i in np.where(dat.trialType == 'cursor')[0] ]), return_inverse = True, axis = 0)
model, ll          = fit_ConditionAveragedModel(model_type, {'n_components' : n_components}, data, conditions)

features_observed  = np.vstack([dat.TX[i] for i in np.where(dat.trialType == 'cursor')[0] ]).astype('float')
features_latent    = model.transform(features_observed)

# Center the data:
day0_latentmeans   = features_latent.mean(axis = 0)
day0_observedmeans = features_observed.mean(axis = 0)
features_latent   -= day0_latentmeans
features_observed -= day0_observedmeans

targets           = np.vstack([dat.targetPos[i] - dat.cursorPos[i] for i in np.where(dat.trialType == 'cursor')[0] ])
```

```python
latent_decoder =  LinearRegression(fit_intercept = False, normalize = False)
full_decoder   =  LinearRegression(fit_intercept = False, normalize = False)

scores_latent   = cross_val_score(latent_decoder, features_latent, targets, cv = 10)
scores_observed = cross_val_score(full_decoder, features_observed, targets, cv = 10)

figSize(7, 7)
plt.scatter(scores_observed, scores_latent)
plt.ylabel('Latent space decoder', fontweight = 'bold')
plt.xlabel('Full decoder', fontweight = 'bold')

minval = np.min([np.min(scores_observed), np.min(scores_latent)]) - 0.05
maxval = np.max([np.max(scores_observed), np.max(scores_latent)]) + 0.05
plt.xlim([minval, maxval])
plt.ylim([minval, maxval])
plt.plot([minval, maxval], [minval, maxval], color = 'k', linestyle = '--')

full_decoder.fit(features_observed, targets)
latent_decoder.fit(features_latent, targets)
```

```python
from copy import deepcopy
from stabilizer import identifyGoodChannels
from scipy.linalg import orthogonal_procrustes

date   = list()
scores = np.zeros((4, 10)) # rows are 0 - no recalibration; 1 - mean recalibration; 2 - stabilizer; 3 - full recalibration
models = list()
Rs     = list()

for i in range(3, 13):
    file        = files[i]
    new_dat     = DataStruct(file)
    date.append(new_dat.date)
        
    features             = np.vstack([new_dat.TX[i] for i in np.where(new_dat.trialType == 'cursor')[0] ])
    targets              = np.vstack([new_dat.targetPos[i] - new_dat.cursorPos[i] for i in np.where(new_dat.trialType == 'cursor')[0] ])
    n_samples, n_chans   = features.shape[0], features.shape[1]
    
    # split dataset in half:
    train_x, test_x     = features[:int(n_samples/2), :], features[int(n_samples/2):, :]
    train_y, test_y     = targets[:int(n_samples/2), :], targets[int(n_samples/2):, :]
    new_observedmeans   = train_x.mean(axis = 0)
 
    # no recalibration:
    scores[0, i-3] = full_decoder.score(test_x - day0_observedmeans, test_y)
    #scores[0, i-3] = np.corrcoef(full_decoder.predict(test_x - day0_observedmeans).T, test_y.T)[[2, 3], [0, 1]].mean()
    
    # mean recalibration:
    scores[1, i-3] = full_decoder.score(test_x - new_observedmeans, test_y)  
    #scores[1, i-3] = np.corrcoef(full_decoder.predict(test_x - new_observedmeans).T, test_y.T)[[2, 3], [0, 1]].mean()


    # Stabilizer:
    coefs_1       = model.components_.T
    
    data          = np.dstack([new_dat.TX[i][:20, :] for i in np.where(new_dat.trialType == 'cursor')[0] ])
    _, conditions = np.unique(np.vstack([new_dat.targetPos[i][0, :] for i in np.where(new_dat.trialType == 'cursor')[0] ]), return_inverse = True, axis = 0)

    new_model, _  = fit_ConditionAveragedModel(model_type, {'n_components' : n_components}, data, conditions)
    coefs_2       = new_model.components_.T
    
    good_chans     = identifyGoodChannels(coefs_1, coefs_2, 120, 0.05)
    R, _           = orthogonal_procrustes(coefs_2[good_chans, :], coefs_1[good_chans, :])
    newdata_latent = new_model.transform(test_x).dot(R)
    scores[2, i-3] = latent_decoder.score(newdata_latent, test_y)
    #scores[2, i-3] = np.corrcoef(latent_decoder.predict(newdata_latent).T, test_y.T)[[2, 3], [0, 1]].mean()
    
    models.append(new_model)
    Rs.append(R)
    
    # Full recalibration:
    lr_FullRecal =  LinearRegression(fit_intercept = False, normalize = False)
    lr_FullRecal.fit(train_x - new_observedmeans, train_y)
    scores[3, i-3] = lr_FullRecal.score(test_x - new_observedmeans, test_y)
    #scores[3, i-3] = np.corrcoef(lr_FullRecal.predict(test_x - new_observedmeans).T, test_y.T)[[2, 3], [0, 1]].mean()

    
    
    
```

```python
figSize(10, 10)

labels = ['No recalibration', 'Mean adaptation', 'Stabilizer', 'Full recalibration']
for i in range(4):
    plt.plot(scores[i, :].T, label = labels[i])
    
plt.xlabel('Time', fontweight = 'bold')
plt.ylabel('R^2', fontweight = 'bold')
plt.xticks(np.arange(len(date)), date, rotation = 40, fontweight = 'bold')
plt.title('Decoder performance across time', fontweight = 'bold')
plt.legend()
```

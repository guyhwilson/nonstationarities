### **Example of subspace realignment in simulator**

We first learn some map $f: X \rightarrow Z \rightarrow Y $ where:

- $X$ is neural activity on day A
- $Y$ is behavior on day A
- $Z$ is some latent space activity on day A

this map is a composition of two linear functions $g: X \rightarrow Z$ (found via linear dimensionality reduction) and $h: Z \rightarrow Y $ (via regression against behavior). On a new day B, we update $g$ while keeping $h$ fixed. To deal with the difference between these subspaces, we now learn a mapping between them via some orthogonal transformation $Q$. Yielding:

$$f': X' \rightarrow Y, f' = h \circ (Q \circ g') $$

We can view this as a form of reduced-rank regression (RRR), where the bottleneck is determined by unsupervised learning instead of the RRR loss function.

#### Setup


First do a sweep of factor analysis latent dimensionality values 

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from CosineTuning import getAngles
import matplotlib.cm as cm

figSize(5, 12)

nUnits         = 192
SNR            = 0.5
nTrainingSteps = 10000


#The first column of initialTuning is the means, and the second two columns are the preferred directions. 
#We make sure that the last two columns are orthogonal (uniformly distributed PDs)
initialTuning       = generateUnits(n_units = nUnits, SNR = SNR)
initialTuning[:, 0] = 0  # for now - assume features are centered 

calVelocity    = np.random.normal(size = (nTrainingSteps, 2))
calNeural      = calVelocity.dot(initialTuning[:,1:].T)  + initialTuning[:, 0].T;  # FR = <velocity, PD> + baseline
calNeural      = calNeural + np.random.normal(size = calNeural.shape) * 0.3      # add gaussian noise


# split into train/test and generate latent trajectories:
train_x, test_x, train_y, test_y = train_test_split(calNeural, calVelocity, train_size = 0.7, random_state = 42)
fa            = FactorAnalysis().fit(train_x)
train_latent  = fa.transform(train_x)
test_latent   = fa.transform(test_x)

# sweep through number of components used and measure holdout R^2:
scores  = np.zeros((20,))
for i in range(20):
    lr        = LinearRegression(fit_intercept = True).fit(train_latent[:,:(i+1)], train_y)
    scores[i] = lr.score(test_latent[:, :(i+1)], test_y)

ref_score = LinearRegression(fit_intercept = True).fit(train_x, train_y).score(test_x, test_y)

# plot performance as function of # components:
plt.subplot(1, 2, 1)
plt.plot(scores[0:], c = 'r')
plt.axhline(ref_score, c = 'k', linestyle = '--')
plt.xlabel("Number of components")
plt.ylabel('R^2')

opt_n = np.argmax(scores) + 1
print('Optimal latent dimensionality: ', opt_n)

# now plot top two weightings and color by angle at timestep:
plt.subplot(1, 2, 2)
angles = getAngles(test_y, returnFullAngle = True)

plt.scatter(test_latent[:, 0], test_latent[:, 1], c = angles, cmap = cm.hsv)
plt.title('Projected neural activity (top two components)')
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Angle (radians)', rotation = 270)
plt.tight_layout()
```

Define code for pulling out matrix mapping to latent space (and test that it's working)

```python
def get_FactorAnalysisMap(fa):
    '''Get transformation matrix that maps observeds into 
       mean latent estimates.
    '''
    Ih    = np.eye(len(fa.components_))
    Wpsi  = fa.components_ / fa.noise_variance_
    cov_z = np.linalg.inv(Ih + np.dot(Wpsi, fa.components_.T))
    
    transform = np.dot(Wpsi.T, cov_z)
    return transform
    

Q = get_FactorAnalysisMap(fa)

x = fa.transform(train_x)
y = (train_x - train_x.mean(axis = 0)).dot(Q)

assert np.linalg.norm(x - y) < 1e-10, print("uh oh")
```

Now fit a subspace-based regression model and test on a given day:

```python
from stabilizer_utils import Stabilizer

model_type = 'FactorAnalysis'
opt_n      = 4

#--------------------
stab         = Stabilizer(model_type = model_type, n_components = opt_n)
stab.fit_ref([train_x])

Q_ref        = get_FactorAnalysisMap(stab.ref_model)
train_latent = (train_x - train_x.mean(axis = 0)).dot(Q_ref)
lr           = LinearRegression(fit_intercept = True).fit(train_latent, train_y)

h      = lr.coef_                                             # map from latent space onto targets
D_coef = h.dot(Q_ref.T)                                       # compose to get neural --> latent --> output 
D      = np.hstack([lr.intercept_[:, np.newaxis], D_coef]).T  # add bias terms

# Normalize the gain of this decoder so that it will output vectors with a magnitude of 1 when the encoded velocity has a magnitude of 1. 
D[:, 0] = D[:,0] / np.linalg.norm(D[1:, :][:, 0]) / np.linalg.norm(initialTuning[:, 1])
D[:, 1] = D[:,1] / np.linalg.norm(D[1:, :][:, 1]) / np.linalg.norm(initialTuning[:, 2])
```

```python
alpha       = 0.94 # amount of exponential smoothing 
delT        = 0.01 # define the time step (10 ms)
nDelaySteps = 20   # define the simulated user's visual feedback delay (200 ms)

import matplotlib.pyplot as plt 
figSize(5,5)

possibleGain = np.linspace(0.5,2.5,10)
nSimSteps    = 50000
meanTTT      = np.zeros((len(possibleGain),))

for g in range(len(possibleGain)):
    print(str(g) + ' / ' + str(len(possibleGain)))
    posTraj, velTraj, rawDecTraj, conTraj, targTraj, neuralTraj, trialStart, ttt = simulateBCIFitts(initialTuning, D,alpha, possibleGain[g], nDelaySteps, delT, nSimSteps)
    meanTTT[g] = np.mean(ttt)

minIdx = np.argmin(meanTTT)
beta   = possibleGain[minIdx]

# Simulate BCI performance with matched neural tuning and decoder, and an optimized gain
PD_shrinkage = 0.5
nSimSteps    = 10000

posTraj, velTraj, rawDecTraj, conTraj, targTraj, neuralTraj, trialStart, ttt = simulateBCIFitts(initialTuning, D, alpha, beta, nDelaySteps, delT, nSimSteps)                                                              
print('With a matched, optimized decoder, mean trial time is ', np.mean(ttt), ' s')

# Simulate a change in neural tuning (specifically, change the PDs only, making sure the total magnitude of tuning is the same)
newTuning = simulateTuningShift(initialTuning, PD_shrinkage = PD_shrinkage)
posTraj_new, velTraj_new, rawDecTraj_new, conTraj_new, targTraj_new, neuralTraj_new, trialStart_new, ttt_new = simulateBCIFitts(newTuning, D, alpha, beta, nDelaySteps, delT, nSimSteps)
print('With changed tuning and a mismatched decoder, the mean trial time is ', np.mean(ttt_new), ' s')

#Control: increase gain in original tuning subspace 
posTraj_control, velTraj_control, rawDecTraj_control, conTraj_control, targTraj_control, neuralTraj_control, trialStart_control, ttt_control = simulateBCIFitts(newTuning, D, alpha, beta/ PD_shrinkage, nDelaySteps, delT, nSimSteps)
print('Control: With changed tuning and a mismatched decoder with restored gain, the mean trial time is ', np.mean(ttt_control), ' s')
```

```python
figSize(5, 15)

plt.subplot(1, 2, 1)
plt.plot(possibleGain, meanTTT, '-o')
plt.xlabel('Gain')
plt.ylabel('Mean Trial Time (s)')
plt.title('Gain sweep', fontweight = 'bold')

minIdx = np.argmin(meanTTT)
beta   = possibleGain[minIdx]

print('Using gain value beta = ', beta)

plt.subplot(1, 2, 2)

from scipy.stats import norm
import seaborn as sns

groups = [times for times in [ttt, ttt_new, ttt_control]]

sns.boxplot(data = groups)
sns.swarmplot(data = groups, color = 'k', alpha = 0.5)
ax = plt.gca()
labels = ['Original','Tuning Change','Gain compensation']
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.ylabel('Trial time (s)')
plt.ylim([0, max([max(i) for i in groups]) + 0.2])
plt.title('Time to target with simulated PD shift', fontweight = 'bold')
```

Now we fit the stabilizer to the changed neural activity:
- finds factor analysis model for new session
- finds an orthogonal transformation that aligns the latent spaces

```python
figSize(5, 7)

B            = 25    # number of stablec channels to identify
thresh       = 0.08  # toss any channels with weight change norm larger than this 

stab.fit_new([neuralTraj_new], B = B, thresh = thresh)


plt.hist(np.linalg.norm(stab.ref_coefs - stab.new_coefs, axis = 1))
plt.axvline(thresh, c = 'k', linestyle = '--', label = 'Cutoff threshold')
plt.xlabel('Loadings change (norm)')
plt.ylabel('Count (# electrodes)')
plt.legend()
```

```python
figSize(7, 22)

angles_old = getAngles(posTraj - targTraj, returnFullAngle = True)
angles_new = getAngles(posTraj_new - targTraj_new, returnFullAngle = True)

old_latent        = stab.ref_model.transform(neuralTraj)
mismatch_latent   = stab.new_model.transform(neuralTraj_new)
stabilized_latent = stab.transform(neuralTraj_new)

plt.subplot(1, 3, 1)
plt.scatter(old_latent[:, 0], old_latent[:, 1], c = angles_old, cmap = cm.hsv)
plt.title('Projected neural activity (reference projection)')
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Angle (radians)', rotation = 270)

plt.subplot(1, 3, 2)
plt.scatter(mismatch_latent[:, 0], mismatch_latent[:, 1], c = angles_new, cmap = cm.hsv)
plt.title('Projected neural activity (new projection)')
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Angle (radians)', rotation = 270)

plt.subplot(1, 3, 3)
plt.scatter(stabilized_latent[:, 0], stabilized_latent[:, 1], c = angles_new, cmap = cm.hsv)
plt.title('Projected neural activity (new projection + stabilizer)')
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Angle (radians)', rotation = 270)
```

Now modify our original decoder to use the new factor analysis model and stabilizing transformation:

```python
G_new     = get_FactorAnalysisMap(stab.new_model)
D_coefnew = h.dot(G_new.dot(stab.R).T)                                # compose to get neural --> latent --> output 
D_new     = np.hstack([lr.intercept_[:, np.newaxis], D_coefnew]).T    # add bias terms

# Normalize the gain of this decoder so that it will output vectors with a magnitude of 1 when the encoded velocity has a magnitude of 1. 
D_new[:, 0] = D_new[:,0] / np.linalg.norm(D_new[1:, :][:, 0]) / np.linalg.norm(newTuning[:, 1])
D_new[:, 1] = D_new[:,1] / np.linalg.norm(D_new[1:, :][:, 1]) / np.linalg.norm(newTuning[:, 2])
```

```python
posTraj_ss, velTraj_ss, rawDecTraj_ss, conTraj_ss, targTraj_ss, neuralTraj_ss, trialStart_ss, ttt_ss = simulateBCIFitts(newTuning, D_new, alpha, beta, nDelaySteps, delT, nSimSteps)
print('With changed tuning and a recalibrated decoder, the mean trial time is ', np.mean(ttt_new), ' s')
```

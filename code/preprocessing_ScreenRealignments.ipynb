**Overview**:

This notebook deals with realigning the cursor and target positions across different tasks, as their values shift for different settings. We perform this realignment by centering all task screens at (0, 0). 

To do so, we loop through all sessions and look at the extremal target values for a given task type (combined across sessions). We then divide the X and Y spread by 2 and store these adjustment values 
in a dictionary *screen_realignments*, where its key value pairs are of the form:

    task_name (str) : values (tuple of floats; x and y adjustments)

Dictionary is saved to /misc_data/screen_realignments.npy

```python
%load_ext autoreload
%autoreload 2

import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from matplotlib import colors 
import glob
import sys
import os

[sys.path.append(f) for f in glob.glob('utils/*')]
from preprocess import DataStruct
from firingrate import raster2FR
from plotting_utils import figSize

files = glob.glob('D:/T5_ClosedLoop/*')


```

Generate example figures of cursor and target positions for two distinct tasks, revealing the positioning problem: 

```python
figSize(7, 15)
dat              = DataStruct(files[10])
n_trials         = 100

cmap             = colors.ListedColormap(['r', 'orange', 'y', 'g', 'k', 'purple', 'violet', 'maroon', 'b'])
unique, targetID = np.unique(np.vstack([dat.targetPos[i][0, :] for i in range(n_trials)]), return_inverse= True, axis = 0)


plt.subplot(1, 2, 1)
plt.xlabel('x pos')
plt.ylabel('y pos')
for i in range(n_trials):
    color       = cmap(targetID[i])
    plt.scatter(dat.targetPos[i][0, 0], dat.targetPos[i][0, 1], color = 'k', s = 200)
    alpha       = [0.3 if targetID[i]== 4 else 1][0]
    plt.plot(dat.cursorPos[i][:, 0], dat.cursorPos[i][:, 1], color = color, alpha = alpha)

plt.subplot(1, 2, 2)
plt.xlabel('x pos')
plt.ylabel('y pos')
for i in range(n_trials):
    trl         = np.where(dat.trialType == 'keyboard')[0][i]
    targPos     = dat.targetPos[trl][0, :]
    plt.scatter(targPos[0], targPos[1], color = 'k', s = 200)
    plt.plot(dat.cursorPos[trl][:, 0], dat.cursorPos[trl][:, 1], color = 'k', alpha = 0.5)
    
plt.suptitle(files[10], fontweight = 'bold')
```

### Generating screen realignments

```python
figSize(15, 15)

task_pos = dict()

# store all task position info in separate dictionary items:
for file in files:
    dat       = DataStruct(file)
    for i, task in enumerate(dat.trialType): 
        if task not in task_pos.keys():
            task_pos[task] = list()   
        task_pos[task].append(dat.targetPos[i][0])

# restructure as np arrays:
for key in task_pos.keys():
    task_pos[key] = np.vstack(task_pos[key])  

    
# visualize the resulting info:
figSize(7, 21)
plt.figure(0)

for i, task in enumerate(task_pos.keys()):
    plt.subplot(1, 3, i + 1)
    coords = task_pos[task]
    plt.scatter(coords[:, 0], coords[:, 1], s = 200, c = 'k')
    plt.title(task)
plt.suptitle('Original coordinates')
    
# now generate alignments and plot the realigned results:
screen_realignments = dict()

for task in task_pos.keys():
    coords = task_pos[task]
    adj    = np.zeros((2,))
    adj[0] = (coords[:, 0].max() - coords[:, 0].min()) / 2 + coords[:, 0].min()
    adj[1] = (coords[:, 1].max() - coords[:, 1].min()) / 2 + coords[:, 1].min()
    
    screen_realignments[task] = adj

plt.figure(1)
#figSize(7, 21)

for i, task in enumerate(task_pos.keys()):
    plt.subplot(1, 3, i + 1)
    coords = task_pos[task]
    adj    = screen_realignments[task]
    plt.scatter(coords[:, 0] - adj[0], coords[:, 1] - adj[1], s = 200, c = 'k')
    plt.title(task)
plt.suptitle('Transformed coordinates')

# Save the data: 
np.save('misc_data/screen_realignments.npy', screen_realignments, allow_pickle = True)
```
